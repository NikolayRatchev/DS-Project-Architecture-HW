Check for missing data

```python
def check_missing(data): 
    
    # Create a DataFrame showing missing counts and percentages
    missing_info = (
        data.isna()
            .sum()
            .reset_index()
            .rename(columns={'index': 'column', 0: 'missing_count'})
    )
    
    # Add missing percentage
    missing_info['missing_percent'] = (missing_info['missing_count'] / len(data) * 100).round(2)
    
    # Filter to only columns with missing values (optional)
    missing_info = missing_info[missing_info['missing_count'] > 0]

    if len(missing_info) == 0: 
        return "no missing vallues in data"
    
    # Sort by number of missing values (descending)
    missing_info = missing_info.sort_values(by='missing_count', ascending=False)
    
    # Display in Jupyter nicely
    return missing_info.reset_index(drop=True)  
```

```python
check_missing(asthma_df)
```
